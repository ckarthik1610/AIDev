```python
def convert_df(all_df):
    new = all_df[["id","agent","user_id","state","created_at","closed_at","merged_at","repo_id"]]
    new = new[new["state"]=="closed"]
    new["status"] = new["merged_at"].apply(lambda x: "Rejected" if pd.isna(x) else "Accepted")
    new = new[["id","agent","user_id","repo_id","status"]]
    return new
```

```python
new_all_df = convert_df(all_pr_df)
new_star_df = convert_df(pr_df)
```
